Question
What makes a song popular?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['track_id'], inplace=True)
df.columns
```

```python
df['duration_min']=df['duration_ms']/60000
df
```

```python
print(df.shape)
print(df['track_genre'].value_counts())
print(df['popularity'].describe())
print(df['explicit'].value_counts())
```

```python
df['popularity_tier']=pd.cut(df['popularity'], bins=[0,25,50,75, 100], labels=['Low', 'Medium', 'High', 'Very High'])
df['length_category']=pd.cut(df['duration_min'], bins=[0, 2, 3.5, 5, 100], labels=['Short', 'Normal', 'Long', 'Very Long'])
print(df['popularity_tier'].value_counts())
print(df['length_category'].value_counts())
```

```python
df.to_csv('spotify_clean.csv', index=False)
```
